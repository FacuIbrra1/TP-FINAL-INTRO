# **Cargo / Leo la tabla:**

```python
# Ruta exacta de la carpeta en tu Drive
folder_path = '/content/drive/MyDrive/ColabNotebooks/Trimestres_anio_archivoTxt'

# Lista para almacenar los DataFrames de cada trimestre
all_dataframes = []

try:
    # Obtenemos la lista de archivos en la carpeta
    file_list = os.listdir(folder_path)
    total_files = len([f for f in file_list if f.endswith('.txt')])
    print(f"Archivos TXT encontrados en la carpeta: {total_files}")
    print("-" * 50)

    # Bucle para leer y cargar cada archivo
    for i, file_name in enumerate(file_list):
        if file_name.endswith('.txt'): # Aseguramos que solo leemos txt
            file_path = os.path.join(folder_path, file_name)

            # Intentamos cargar con el separador típico de EPH (punto y coma)
            try:
                temp_df = pd.read_csv(file_path, sep=';', encoding='latin-1')
                all_dataframes.append(temp_df)
                print(f"Cargado {i+1}/{total_files}: {file_name}")
            except Exception:
                # Si falla, intentamos con separador de coma
                try:
                    temp_df = pd.read_csv(file_path, sep=',', encoding='latin-1')
                    all_dataframes.append(temp_df)
                    print(f"Cargado {i+1}/{total_files}: {file_name} (Separador ,)")
                except Exception as e:
                    print(f"⚠️ Error al cargar {file_name}. Se saltará este archivo.")

    # Concatenamos todos los DataFrames cargados
    if all_dataframes:
        df_total = pd.concat(all_dataframes, ignore_index=True)
        print("\n" + "="*50)
        print("✅ ¡Bases de datos concatenadas con éxito!")
        print(f"Dimensiones de la base total (filas, columnas): {df_total.shape}")
        # Agregamos una variable PERIODO clave para el análisis de series de tiempo
        df_total['PERIODO'] = df_total['ANO4'].astype(str) + 'Q' + df_total['TRIMESTRE'].astype(str)
        print("Primeras filas de la base total:")
        display(df_total.head())
    else:
        print("❌ ERROR: No se pudo cargar ningún archivo CSV. Revisa la ruta y los archivos.")

except FileNotFoundError:
    print(f"❌ ERROR: La ruta {folder_path} no existe. Revisa que el nombre de la carpeta sea exacto.")
```

```python
# Convertimos todos los nombres de columnas a minúsculas y eliminamos espacios
df_total.columns = df_total.columns.str.lower().str.strip()

print("✅ Nombres de columnas estandarizados a minúsculas y sin espacios.")
print("-" * 50)
print("Columnas clave disponibles:")
print(df_total.filter(items=['p21', 'pp3e_tot', 'estado', 'aglomerado', 'periodo']).columns.tolist())
```

```python
# Aglomerados elegidos: 34 (Mar del Plata - Batán) y 12 (Corrientes)
AGLOMERADO_1 = 34
AGLOMERADO_2 = 12

# 1. Filtrar solo los aglomerados de interés
# Convertimos 'aglomerado' a numérico
df_total['aglomerado'] = pd.to_numeric(df_total['aglomerado'], errors='coerce')
df_filtrado_final = df_total[df_total['aglomerado'].isin([AGLOMERADO_1, AGLOMERADO_2])].copy()

print("✅ Paso 1.B: Filtrado de Aglomerados completado.")
print(f"Registros iniciales de {AGLOMERADO_1} y {AGLOMERADO_2}: {len(df_filtrado_final)}")
print("-" * 50)


# 2. CREACIÓN ROBUSTA DE INGRESO_HORARIO
# Paso 2.1: Limpiar y Forzar conversión a Numérico (FLOAT)

# Limpiamos espacios y convertimos a FLOAT. Esto resuelve el error TypeError 'int' and 'str'.
for col in ['p21', 'pp3e_tot']:
    # 1. Aseguramos que es texto y quitamos espacios (esencial para TXT)
    df_filtrado_final[col] = df_filtrado_final[col].astype(str).str.strip()
    # 2. Convertimos a numérico: si hay un error (no es número), pone NaN
    df_filtrado_final[col] = pd.to_numeric(df_filtrado_final[col], errors='coerce')


# Paso 2.2: Calcular Ingreso Horario
# Horas 0 se convierten a NaN para evitar la división por cero.
df_filtrado_final['pp3e_tot'] = df_filtrado_final['pp3e_tot'].replace(0, np.nan)
df_filtrado_final['INGRESO_HORARIO'] = df_filtrado_final['p21'] / df_filtrado_final['pp3e_tot']

print("✅ Creación de Ingreso Horario completada.")
print("-" * 50)
print("Tipos de datos de las variables clave (deben ser 'float64'):")
print(df_filtrado_final[['p21', 'pp3e_tot', 'INGRESO_HORARIO']].dtypes)


# 3. Filtrar solo a la población OCUPADA con ingresos válidos (ESTADO=1)
# Convertimos 'estado' a numérico
df_filtrado_final['estado'] = pd.to_numeric(df_filtrado_final['estado'], errors='coerce')

df_limpio_final = df_filtrado_final[
    (df_filtrado_final['estado'] == 1) &
    (df_filtrado_final['INGRESO_HORARIO'].notna()) &
    (df_filtrado_final['INGRESO_HORARIO'] > 0)
].copy()

print("\n✅ Limpieza Final (Filtro Ocupados con Ingreso Válido) completada.")
print(f"Total de registros limpios para el análisis de ingresos: {len(df_limpio_final)}")
```
